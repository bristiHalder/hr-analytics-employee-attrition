## Data Cleaning and Preparation

```python
#checking missing values
df.isnull().sum()
```

```python
#Convert Attrition to Binary
df['Attrition_Flag'] = df['Attrition'].map({'Yes': 1, 'No': 0})

df[['Attrition', 'Attrition_Flag']].head()
```

```python
# Drop Unnecessary Columns
drop_cols = [
    'EmployeeCount',
    'EmployeeNumber',
    'Over18',
    'StandardHours'
]

df = df.drop(columns=drop_cols)
```

```python
# Check Categorical vs Numerical Columns
categorical_cols = df.select_dtypes(include='object').columns
numerical_cols = df.select_dtypes(include=np.number).columns

categorical_cols, numerical_cols
```

```python
#Create Experience Buckets
df['Experience_Group'] = pd.cut(
    df['YearsAtCompany'],
    bins=[0, 2, 5, 10, 40],
    labels=['0-2 years', '3-5 years', '6-10 years', '10+ years']
)

df[['YearsAtCompany', 'Experience_Group']].head()
```

## Exploratory Data Analysis (EDA)

```python
# Overall Attrition Rate (KPI)
attrition_rate = df['Attrition_Flag'].mean() * 100
attrition_rate
```

```python
print(f"Overall Attrition Rate: {attrition_rate:.2f}%")
```

```python
# Attrition by Department
dept_attrition = df.groupby('Department')['Attrition_Flag'].mean().sort_values(ascending=False) * 100
dept_attrition
```

```python
plt.figure(figsize=(6,4))
sns.barplot(x=dept_attrition.values, y=dept_attrition.index)
plt.title('Attrition Rate by Department')
plt.xlabel('Attrition Rate (%)')
plt.ylabel('Department')
plt.show()
```

```python
# Overtime vs Attrition
overtime_attrition = df.groupby('OverTime')['Attrition_Flag'].mean() * 100
overtime_attrition
```

```python
plt.figure(figsize=(4,4))
sns.barplot(x=overtime_attrition.index, y=overtime_attrition.values)
plt.title('Attrition Rate by Overtime')
plt.ylabel('Attrition Rate (%)')
plt.show()
```

```python
# Attrition by Experience Group
exp_attrition = df.groupby('Experience_Group')['Attrition_Flag'].mean() * 100
exp_attrition
```

```python
plt.figure(figsize=(6,4))
sns.barplot(x=exp_attrition.index, y=exp_attrition.values)
plt.title('Attrition Rate by Experience Group')
plt.ylabel('Attrition Rate (%)')
plt.xlabel('Experience Group')
plt.show()
```

```python
# Income vs Attrition
plt.figure(figsize=(6,4))
sns.boxplot(x='Attrition', y='MonthlyIncome', data=df)
plt.title('Monthly Income vs Attrition')
plt.show()
```

```python
# Clean Column Names
# Make column names SQL-friendly
df.columns = (
    df.columns
      .str.strip()
      .str.lower()
      .str.replace(' ', '_')
      .str.replace('-', '_')
)

df.columns
```

```python
# Select Final Columns
final_cols = [
    'age',
    'gender',
    'department',
    'jobrole',
    'monthlyincome',
    'yearsatcompany',
    'experience_group',
    'overtime',
    'attrition',
    'attrition_flag'
]

df_final = df[final_cols]
df_final.head()
```

```python
# Export Cleaned Dataset
df_final.to_csv('cleaned_hr_attrition_data.csv', index=False)
```
